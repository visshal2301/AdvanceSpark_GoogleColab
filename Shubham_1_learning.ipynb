```python
df_transactions_filter=df_transactions_filter.where(col("amt")>=500).show()

```

```python
# case statement
from pyspark.sql.functions import when, col
df_transactions=df_transactions.withColumn("amt_category" ,
              when(col("amt").cast("double") < 500,"500_less")
              .when((col("amt").cast("double") >= 500) & (col("amt").cast("double") < 1000),"1000_less")
              .otherwise("1000_more"))
df_transactions.show()
```

```python
# Case Statement With Expr
from pyspark.sql.functions import expr

df_transactions = df_transactions.withColumn(
    "amt_category",
    expr(
        """
        CASE
            WHEN CAST(amt AS DOUBLE) < 500 THEN '500_less'
            WHEN CAST(amt AS DOUBLE) >= 500 AND CAST(amt AS DOUBLE) < 1000 THEN '1000_less'
            ELSE '1000_More'
        END
        """
    )
)

df_transactions.show()
```

```python
from pyspark.sql.functions import to_date


df_transactions=df_transactions.withColumn("start_date",to_date(col("start_date"),'yyyy-MM-dd'))
df_transactions.show()
df_transactions.printSchema()
```

```python
#union
#df_transactions_union=df_transactions.union(df_transactions)
df_transactions_union=df_transactions.unionByName(df_transactions)
print(f" union count :- {df_transactions_union.count()}")
print(f" union count :- {df_transactions.count()}")
```

```python
# Sorting

from pyspark.sql.functions import asc,desc
import pyspark.sql.functions as F

df_transactions_agg=df_transactions.groupBy("cust_id","expense_type").agg(F.count("*").alias("total_cnt"),F.sum("amt").alias("total_amt"))
df_transactions_agg.orderBy(col("cust_id").asc()).show()
```

```python
## Unique
## select distinct cust_id from df_transactions
df_transactions_agg.persist()
df_cust_id=df_transactions.select("cust_id").distinct().show()
```

```python
#windows Function
#select *,max("amt") over (partition by expense_type order by  amt desc) as Max_amt from df_transactions

from pyspark.sql.window import Window
import pyspark.sql.functions as F

windows_spec=Window.partitionBy(F.col("expense_type")).orderBy(F.col("total_amt").desc())
max_func=F.max(F.col("total_amt")).over(windows_spec)

df_new=df_transactions_agg.withColumn("rn",max_func).show()
```
